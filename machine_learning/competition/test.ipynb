```python
from sklearn.model_selection import KFold
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# 製造假資料
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)

# K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]
    
    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)
    score = model.score(X_val, y_val)
    scores.append(score)
    print(f"Fold {fold+1}: {score:.4f}")

print(f"\n平均準確率: {np.mean(scores):.4f}")
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 標準化（PCA 和 K-Means 都需要）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 降維到 2D（方便視覺化）
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(f"原始維度: {X.shape[1]} → 降維後: {X_pca.shape[1]}")
print(f"解釋了多少變異: {pca.explained_variance_ratio_.sum():.2%}")

# K-Means 分成3群
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_pca)

# 畫圖
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Set1', alpha=0.6)
plt.title("PCA 降維 + K-Means 分群結果")
plt.xlabel("第一主成分")
plt.ylabel("第二主成分")
plt.colorbar(label='群別')
plt.show()
```
