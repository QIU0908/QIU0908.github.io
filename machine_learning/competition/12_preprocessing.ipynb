```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

data = {
    '年齡': [25, np.nan, 32, 28, 35, np.nan, 29],
    '收入': [50000, 80000, np.nan, 60000, np.nan, 70000, 55000],
    '性別': ['男', '女', '男', np.nan, '女', '男', '女'],
    '是否購買': ['是', '否', '是', '否', '是', '否', '是']
}

df = pd.DataFrame(data)
print(df)
print(f"\n缺失值數量:\n{df.isnull().sum()}")
```

```python
# 數值欄位用平均值填補
imputer = SimpleImputer(strategy='mean')
df[['年齡', '收入']] = imputer.fit_transform(df[['年齡', '收入']])

# 類別欄位用最常出現的值填補
imputer_cat = SimpleImputer(strategy='most_frequent')
df[['性別']] = imputer_cat.fit_transform(df[['性別']])

print(df)
print(f"\n缺失值數量:\n{df.isnull().sum()}")
```

```python
# 性別：男=0, 女=1
df['性別'] = df['性別'].map({'男': 0, '女': 1})

# 是否購買：否=0, 是=1
df['是否購買'] = df['是否購買'].map({'否': 0, '是': 1})

print(df)
```

```python
X = df[['年齡', '收入', '性別']].values
y = df['是否購買'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("標準化前:")
print(X[:3])
print("\n標準化後:")
print(X_scaled[:3].round(2))
```
