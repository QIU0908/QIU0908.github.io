K-Means → 把資料分群  
  競賽用法：  
  把群別當新特徵加入模型  
  例如：客戶分成4群後  
  「屬於第幾群」變成一個新特徵  

PCA → 把很多特徵壓縮成少數特徵  
  競賽用法：  
  100個特徵 → 壓縮成10個主成分  
  減少過擬合，加快訓練  

DBSCAN → 找異常點  
  競賽用法：  
  標記出異常資料  
  清理資料或當作新特徵  

需要標準化：Linear, Logistic, SVM, PCA, K-Means  
不需要標準化：Decision Tree, Random Forest, XGBoost, LightGBM  

| 問題類型       | 資料規模     | 首選模型                  | 次選模型             | 重點考量                  |
|----------------|--------------|---------------------------|----------------------|---------------------------|
| **回歸**       | 小～中       | Linear Regression / Ridge | XGBoost              | 線性關係優先              |
| **回歸**       | 中～大       | XGBoost / LightGBM        | Random Forest        | 性能優先                  |
| **分類**       | 小           | Logistic Regression / SVM | Random Forest        | 可解釋性                  |
| **分類**       | 中～大       | XGBoost                   | LightGBM             | 準確度 + 速度             |
| **超大資料**   | 10萬筆以上   | LightGBM                  | XGBoost              | 訓練速度與記憶體          |
| **極需解釋**   | 任何規模     | Decision Tree + SHAP      | Logistic Regression  | 業務理解與可解釋性        |
| **分群**       | -            | K-Means                   | DBSCAN               | 群形狀與噪音處理          |
| **高維資料**   | -            | PCA + 其他模型            | SVM                  | 先降維再建模              |
